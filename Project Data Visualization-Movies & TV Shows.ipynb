# Soal no 3
 Average Rotten Tomatoes rating per directors 

```python
df3
```

```python
df_director = df3['Directors'].str.get_dummies(',')
df_director.head()
```

```python
df_director['Sutradara'] = df_director.idxmax(1)
df_director
```

```python
df_sutradara = pd.merge(df3, df_director['Sutradara'], left_index=True, right_index=True)
df_sutradara.head(20)
```

```python
df_sutradara.info()
```

```python
df_sutradara['Rating'] = df_sutradara['Rotten Tomatoes'].str[:2]
df_baru = df_sutradara.head(10)
df_baru
```

```python
df_sutradara.isnull().sum()
```

```python
df_baru1 = df_baru.astype({'Rating' : 'int'})
df_baru1
```

```python
df_baru1.info()
```

```python
dfr = df_baru1.groupby(['Sutradara'])['Rating'].mean().reset_index(name='Rating')
dfr
```

```python
ax = sns.barplot(x='Rating', y='Sutradara',
                  data=dfr,
                 errwidth=0)

for i in ax.containers:
    ax.bar_label(i,)
```
